```python
shelves_freq = Counter()

for file in files:
    book_data = open(file,encoding='utf-8')
    shelves = [shelf.strip() for shelf in book_data.readlines()[3:] if shelf.strip() not in shelves_to_ignore]
    shelves_freq.update(shelves)
```

```python
x_axis = []
y_axis = []

sns.set_style("darkgrid")
for shelf,count in shelves_freq.most_common(50):
    y_axis.append(shelf) 
    x_axis.append(count)

fig = plt.figure( figsize=( 5 , 10 ) )


graph = sns.barplot( x=x_axis, y=y_axis, dodge=False , color = '#3242a8' )

graph.set_title('Goodreads shelves of banned books' , size = 20) 
graph.set_xlabel('Number of books' , size = 14) 
graph.set_ylabel('Shelf' , size = 14 )

plt.xticks(rotation= 90)


plt.show()
```

```python
data = []

for file in files:
    row = dict()
    book_data = open(file,encoding='utf-8').readlines()
    title = book_data[2].strip()
    shelves = [shelf.strip() for shelf in book_data[3:] if shelf.strip() not in shelves_to_ignore]
    row['title'] = title
    row['shelves'] = shelves
    if len(shelves)>0:
        data.append(row)
    
json_data = json.dumps(data)
with open('books_shelves.json','w',encoding='utf-8') as out:
    out.write(json_data)
```

```python
selected_shelves = []
ignore_list = ['black-bookcase','holly-black','blacksmith','black-dagger-brotherhood',
               'black-dagger-brotherhood-series',
'black-dagger','pretty-brown-eyes-48','black-swan-financial']

for shelf in shelves_freq.keys():
    if re.search('(lgbt)|(queer)|(gay)|(black)|(hispanic)',shelf):
        if shelf not in ignore_list:
            selected_shelves.append(shelf)
```

```python
out = open('lgbtq_black_hispanic.csv','w',encoding='utf-8')

count = 0 
for file in files:
    book_data = open(file,encoding='utf-8').readlines()
    title = book_data[2].strip()
    title = re.sub(',', ' ', title)
    shelves = [shelf.strip() for shelf in book_data[3:] if shelf.strip() not in shelves_to_ignore]
    for shelf in shelves:
        if shelf in selected_shelves:
            category = 'LGBTQ'
            if re.search('black',shelf,re.IGNORECASE):
                category = 'Black'
            if re.search('hispanic',shelf,re.IGNORECASE):
                category = 'Hispanic'
            out.write(f"\"{title}\",{shelf},{category}\n")
            break
out.close()
```
